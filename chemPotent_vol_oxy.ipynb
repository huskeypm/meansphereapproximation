### After jobs have finished, load them via this section

```python
import cPickle as pickle

def GetorRunData(ionIndex,key,noRun=False):
    m = np.shape(volumes)[0]
    n = np.shape(oxys)[0]
    matrixData = np.zeros([m,n])

    params = runner.Params()
    muiexsPrev=0
    psiPrev=0
    for i,vol in enumerate(volumes):
       for j,nOxy in enumerate(oxys):
         params.V_i = vol
         #print "TROUBLESHOOTING"   
         #params.nOxy = 5 # oxy
         params.nOxy = nOxy #    
         params.pklName = "%f_%f_test.pkl"%(vol,nOxy)
         params.setup()   
        
         daFile = runner.daIter(params=params, muiexsPrev=muiexsPrev,psiPrev=psiPrev, noRun=noRun)
         #print daFile
         #break    
         pkl_file = open(daFile, 'rb')#Ca
         results = pickle.load(pkl_file)
         pkl_file.close()

         muiexsPrev=results['muFilter']
         psiPrev=results['donnanPotential'] 
         #print muiexsPrev

         data = results[key]   
         matrixData[i,j] = data[ ionIndex ]   

         #print daFile, mVolOxy[i,j]   
    return matrixData
```
